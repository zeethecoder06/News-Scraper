```python
url="https://www.indiatoday.in/top-stories"
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_soup(next_url):
    base_url="https://www.indiatoday.in"
    url = base_url+next_url
    try:
        page=requests.get(url)
        if page.status_code==200:
            print("page load")
            return BeautifulSoup(page.text,'lxml')
        else:
            print("error",page.status_code)
    except Exception as e:
        print("it's an error")
        print(e)
```

```python
def get_news(soup):
    news_box = soup.find('div',{'class':'view-content'})
    all_news = news_box.find_all('div',attrs={'class':'catagory-listing'})
    dataset =[]
    for news in all_news:
        try:
            heading = news.h2.text
            detail = news.p.text
            link = news.h2.a.attrs.get('href')
            dataset.append({
                'headline':heading,
                'detail':detail,
                'link':link
            })
        except Exception as e:
            pass

    try:
        next = soup.find('li',attrs={'class':'pager-next'}).a.attrs.get('href')
        return {'dataset':dataset,'next':next} 
    except:
        return {'dataset':dataset,'next':None}
```

```python
## initial call
try:
    page=requests.get(url)
    if page.status_code==200:
        print("page load")
    else:
        print("error",page.status_code)
except Exception as e:
    print(e)
    print("it's an error")
soup=BeautifulSoup(page.text,'lxml')
```

```python
bigger_dataset=[]
while True:
    output = get_news(soup)
    if output.get('dataset'):
        bigger_dataset.extend(output.get('dataset'))
    if output.get('next'):
        soup = get_soup(output.get('next'))
    else:
        break
```

```python
import pandas as pd
df =pd.DataFrame(bigger_dataset)
df.to_csv('top_news.csv')
df        
```
